```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *

# Create Spark Session
spark = SparkSession.builder.appName("EmployeeDataFrameAssignment").getOrCreate()

# Employee Data
data = [
    (101, "John", "IT", 50000, "Chennai", 30),
    (102, "Alice", "HR", 60000, "Bangalore", 28),
    (103, "Bob", "Finance", 70000, "Mumbai", 35),
    (104, "David", "IT", 80000, "Chennai", 40),
    (105, "Eva", "HR", 55000, "Hyderabad", 25),
    (106, "Frank", "Finance", 90000, "Mumbai", 45)
]

columns = ["emp_id", "emp_name", "department", "salary", "city", "age"]

emp_df = spark.createDataFrame(data, columns)

emp_df.show()
```

```python
# Section 7 – Null Handling
# Create Sample Data with NULL Salary
```

```python
data_null = [
    (101, "John", "IT", 50000, "Chennai", 30),
    (102, "Alice", "HR", None, "Bangalore", 28),
    (103, "Bob", "Finance", 70000, "Mumbai", 35),
    (104, "David", "IT", None, "Chennai", 40)
]

df_null = spark.createDataFrame(
    data_null,
    columns
)
```

```python
# 1. Display NULL salary records
```

```python
df_null.filter(
    col("salary").isNull()
).show()
```

```python
# 2. Replace NULL salary with 50000
```

```python
df_null.fillna(
    {"salary":50000}
).show()
```

```python
# 3. Drop records containing NULL values
```

```python
df_null.na.drop().show()
```

```python
# 4. Count NULL salary records
```

```python
df_null.filter(
    col("salary").isNull()
).count()
```

```python
# 5. Replace NULL salary with Department Average Salary
```

```python
dept_avg = df_null.groupBy("department").agg(
    avg("salary").alias("avg_salary")
)

result = df_null.join(
    dept_avg,
    "department",
    "left"
).withColumn(
    "salary",
    when(col("salary").isNull(), col("avg_salary"))
    .otherwise(col("salary"))
).drop("avg_salary")

result.show()
```
