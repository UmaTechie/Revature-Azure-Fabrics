```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("RDDAssignment-2").getOrCreate()
sc = spark.sparkContext

emp_rdd = sc.parallelize([
    (101,"John","IT",50000),
    (102,"Alice","HR",60000),
    (103,"Bob","IT",70000),
    (104,"David","Finance",80000),
    (105,"Eva","HR",55000),
    (106,"Frank","IT",65000),
    (107,"Grace","Finance",90000),
    (108,"Helen","HR",62000)
])
print(emp_rdd.collect())
```

```python
# Display only employee names using map() and collect().
names = emp_rdd.map(lambda x: x[1]).collect()
print(names)
```

```python
# Collect all Finance department employees.
finance_emp = emp_rdd.filter(lambda x: x[2] == "Finance").collect()
print(finance_emp)
```

```python
# Collect all Finance department employees.
finance_emp = emp_rdd.filter(lambda x: x[2] == "Finance").collect()
print(finance_emp)
```

```python
# Collect employees earning more than 70000.
high_salary = emp_rdd.filter(lambda x: x[3] > 70000).collect()
print(high_salary)
```

```python
# Collect all unique department names.
departments = emp_rdd.map(lambda x: x[2]).distinct().collect()
print(departments)
```

```python
# Collect employee IDs and salaries as tuples.
id_salary = emp_rdd.map(lambda x: (x[0], x[3])).collect()
print(id_salary)
```

```python
# Collect employees sorted by salary.
sorted_salary = emp_rdd.sortBy(lambda x: x[3]).collect()
print(sorted_salary)
```

```python
# Collect only HR employee names.
hr_names = emp_rdd.filter(lambda x: x[2] == "HR") \
                  .map(lambda x: x[1]) \
                  .collect()
print(hr_names)
```

```python
# Collect all records after increasing salary by 10%.
updated_salary = emp_rdd.map(
    lambda x: (x[0], x[1], x[2], x[3] * 1.10)
).collect()
print(updated_salary)
```

```python
# Count employees after removing duplicates.
employee_count = emp_rdd.distinct().count()
print(employee_count)
```

```python
# Count employees with salary between 50000 and 70000.
count_emp = emp_rdd.filter(
    lambda x: 50000 <= x[3] <= 70000
).count()
print(count_emp)
```

```python
# Retrieve the first employee earning above 60000.
first_emp = emp_rdd.filter(
    lambda x: x[3] > 60000
).first()
print(first_emp)
```

```python
# Find highest salary.
highest_salary = emp_rdd.map(lambda x: x[3]).max()
print(highest_salary)
```
