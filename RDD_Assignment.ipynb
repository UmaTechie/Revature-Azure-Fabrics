```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("RDDAssignment").getOrCreate()
sc = spark.sparkContext

emp_rdd = sc.parallelize([
    (101,"John","IT",50000),
    (102,"Alice","HR",60000),
    (103,"Bob","IT",70000),
    (104,"David","Finance",80000),
    (105,"Eva","HR",55000),
    (106,"Frank","IT",65000),
    (107,"Grace","Finance",90000),
    (108,"Helen","HR",62000)
])
print(emp_rdd.collect())
```

```python
# 1. Use map() to extract employee names.
names_rdd = emp_rdd.map(lambda x: x[1])
print(names_rdd.collect())
```

```python
# 2. Use map() to create (emp_id, salary) pairs.
id_salary_rdd = emp_rdd.map(lambda x: (x[0], x[3]))
print(id_salary_rdd.collect())
```

```python
# 3. Use filter() to get employees with salary > 60000.
high_salary_rdd = emp_rdd.filter(lambda x: x[3] > 60000)
print(high_salary_rdd.collect())
```

```python
# 4. Use filter() to get only IT department employees.
dept_emp = emp_rdd.filter(lambda x: x[2] == "IT")
print(dept_emp.collect())
```

```python
# 5. Use distinct() to find unique departments.
dept_rdd = emp_rdd.map(lambda x: x[2])
unique_dept = dept_rdd.distinct()
print(unique_dept.collect())
```

```python
# 6. Create another RDD and perform union().
new_emp_rdd = sc.parallelize([
    (109,"Kiranmayi","IT",80000)])
union_rdd = emp_rdd.union(new_emp_rdd)
print(union_rdd.collect())
```

```python
# 7. Perform intersection() between two employee ID RDDs.
rdd1 = sc.parallelize([101,102,103,104,105])
rdd2 = sc.parallelize([103,104,105,106,107])
common_ids = rdd1.intersection(rdd2)
print(common_ids.collect())
```

```python
# 8. Use flatMap() to split department names into characters.
dept_char_rdd = emp_rdd.map(lambda x: x[2]).flatMap(lambda x: list(x))
print(dept_char_rdd.collect())
```

```python
# 9. Use mapValues() on Pair RDD to increase salary by 10%.
salary_pair_rdd = emp_rdd.map(lambda x: (x[0], x[3]))
updated_salary_rdd = salary_pair_rdd.mapValues(lambda salary: salary * 1.10)
print(updated_salary_rdd.collect())
```

```python
# 10. Use reduceByKey() to calculate total salary by department.
dept_salary_rdd = emp_rdd.map(lambda x: (x[2], x[3]))
total_salary_rdd = dept_salary_rdd.reduceByKey(lambda a, b: a + b)
print(total_salary_rdd.collect())
```
