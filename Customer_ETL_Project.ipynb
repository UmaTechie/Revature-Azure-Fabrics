```python
customer_df["Age"] = customer_df["Age"].fillna(
    customer_df["Age"].mean())
customer_df["TotalPurchase"] = customer_df["TotalPurchase"].fillna(0)
```

```python
customer_df.shape
```

```python
customer_df = customer_df.drop_duplicates()
customer_df.shape
```

```python
customer_df["CustomerID"] = customer_df["CustomerID"].astype(int)
customer_df["Age"] = customer_df["Age"].astype(int)
customer_df["JoinDate"] = pd.to_datetime(
customer_df["JoinDate"])
```

```python
engine = create_engine(
    "mysql+pymysql://root:mysql@127.0.0.1:3306/customer_etl")
engine
```

```python
from sqlalchemy import text

with engine.connect() as conn:
    result = conn.execute(text("SELECT DATABASE();"))
    print(result.fetchone())
```

```python
# 4.Find top customers by purchase.
customer_df.to_sql(
    name="customers_fact",
    con=engine,
    if_exists="replace",
    index=False)
print("Customers Loaded")
```

```python
# 5.Analyze city-wise revenue
pd.read_sql("""SELECT City, SUM(TotalPurchase) Revenue FROM customers_fact
GROUP BY City ORDER BY Revenue DESC""",con=engine)
```

```python
# 6.Rank customers using window functions
pd.read_sql("""SELECT CustomerName, TotalPurchase,
RANK() OVER(ORDER BY TotalPurchase DESC) AS CustomerRank FROM customers_fact""",con=engine)
```

```python
# 7. Total Customers
pd.read_sql("""SELECT COUNT(*) AS TotalCustomers FROM customers_fact""",con=engine)
```

```python
# 8. City-wise Customer Count
pd.read_sql("""SELECT City, COUNT(*) AS CustomerCount FROM customers_fact GROUP BY City""",con=engine)
```

```python
# 9. City-wise Revenue
pd.read_sql("""SELECT City, SUM(TotalPurchase) AS Revenue FROM customers_fact GROUP BY City""",con=engine)
```

```python
# 10. fetch Above Average Customers information
pd.read_sql("""SELECT *FROM customers_fact WHERE TotalPurchase >
(SELECT AVG(TotalPurchase) FROM customers_fact)""",con=engine)
```

```python
# 11. find Customer Acquisition Trend
pd.read_sql("""SELECT MONTH(JoinDate) AS MonthNo,
COUNT(*) AS NewCustomers FROM customers_fact GROUP BY MONTH(JoinDate) ORDER BY MonthNo""",con=engine)
```
